#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code
# o calculate the standard deviation of those numbers:

# 1. Work out the Mean (the simple average of the numbers)
# 2. Then for each number: subtract the Mean and square the result
# 3. Then work out the mean of those squared differences.
# 4. Take the square root of that and we are done!
# The formula actually says all of that, and I will show you 
#st = [9, 2, 5, 4, 12, 7]
def stand_dev(col):
    mean_count = 0
    number_n = 0
    for n in col:
        mean_count += n
        #print(mean_count)
        number_n += 1
        #print(number_n)
        avg_n = mean_count / number_n
        #print(avg_n)
    new_mean_count = 0
    for n in col:
        new_n = (n - avg_n)**2
        #print(new_n)
        new_mean_count += new_n
        new_mean = new_mean_count/ (number_n -1)
    return (new_mean **.5)

#stand_dev(lst)
#https://www.mathsisfun.com/data/standard-deviation-formulas.html - helped me understand std AND go step by step to code it!
````

*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`
